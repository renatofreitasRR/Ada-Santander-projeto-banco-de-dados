```python
import pandas as pd

df_produtos = pd.read_csv('produtos.csv')
df_vendas = pd.read_csv('vendas.csv')
```

```python
df_vendas.head()
```

```python
df_produtos.head()
```

```python
#Formatando valor de data das vendas
df_vendas["Date"] = pd.to_datetime(df_vendas["Date"])
```

```python
df_vendas.head()
```

```python
#Formatando valor monetário dos preços dos produtos
df_produtos["Preco"] = df_produtos["Preco"].apply(lambda x: x[1:])
```

```python
df_produtos.head()
```

```python
#Salvando novos CSV's
df_produtos.to_csv('produtos_formatados.csv', index=False)
df_vendas.to_csv('vendas_formatadas.csv', index=False)
```
